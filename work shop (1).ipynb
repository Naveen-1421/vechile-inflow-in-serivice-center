```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error

data4['inputDates'] = pd.to_datetime(data4['inputDates'])

data4['inputDates'] = pd.to_datetime(data4['inputDates'])

data4['year'] =data4['inputDates'].dt.year
data4['month'] = data4['inputDates'].dt.month
data4['day'] = data4['inputDates'].dt.day
data4['hour'] = data4['inputDates'].dt.hour

encoder = OneHotEncoder(handle_unknown='ignore')
dayOfWeek_encoded = encoder.fit_transform(data4[['dayOfWeek']])
dayOfWeek_df = pd.DataFrame(dayOfWeek_encoded.toarray(), columns=encoder.get_feature_names_out(['dayOfWeek']))

X =pd.concat([data4[['year', 'month', 'day', 'hour']], dayOfWeek_df], axis=1)

y = data4["Hour"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

```
